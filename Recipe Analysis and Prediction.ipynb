### **Outlier Treatment for Nutritional Attributes**
- Focused on key nutritional columns: `calories`, `protein`, `fat`, and `sodium`

```python
dfy = df_filtered[['calories', 'protein',	'fat',	'sodium']]

Q1 = dfy.quantile(0.25)
Q3 = dfy.quantile(0.75)
IQR = Q3 - Q1

outliers = ((dfy < (Q1 - 1.5 * IQR)) |(dfy > (Q3 + 1.5 * IQR))).any(axis=1)
outliers_indices = dfy[outliers].index

outliers_count = outliers.sum()
# outliers_count

plt.figure(figsize=(20,10))
for i,column in enumerate(dfy.columns,1):
    plt.subplot(2,4,i)
    sns.boxplot(data=dfy[column])
    plt.title(column)
plt.tight_layout()
plt.show()
```

**Outlier Treatment: IQR Method**
To handle outliers in our dataset, we employed the Interquartile Range (IQR) method:

**Boundary Calculation:**

    We used 1.5 times the IQR to set upper and lower boundaries.
    Lower Bound = Q1 - (1.5 * IQR)
    Upper Bound = Q3 + (1.5 * IQR)


**Outlier Identification:**

    Data points falling outside these boundaries were flagged as outliers.


**Treatment:**

    Outliers were capped at the boundary values.

```python
for feature in Q1.index:
    LB = Q1[feature] - 1.5 * IQR[feature]
    UB = Q3[feature] + 1.5 * IQR[feature]
    dfy[feature] = dfy[feature].clip(lower=LB, upper=UB)

plt.figure(figsize=(20,10))
for i,column in enumerate(dfy.columns,1):
    plt.subplot(2,4,i)
    sns.boxplot(data=dfy[column])
    plt.title(column)
plt.tight_layout()
plt.show()
```

### **Recipe Information Enhancement**
- **Action**: Added recipe information to enrich the dataset

```python
# Recipe directions and ingrdients extraction function

def get_recipe_info(recipe_title):
    """
    Takes a recipe title and returns a list of 
    recipe directions length and ingrdients length for the recipe.

    Args:
        recipe_title (_type_): _description_

    Returns:
        _type_: _description_
    """
    recipe_dic = [i for i in full_recipe if "title" in i.keys() and i["title"]==recipe_title]
    if len(recipe_dic) == 0:
        return pd.Series({"len_directions":None, "len_ingredients":None})
    else:
        recipe_dic = recipe_dic[0]
    len_directions = len(recipe_dic["directions"])
    len_ingredients = len(recipe_dic["ingredients"])
    return pd.Series({"len_directions":len_directions, "len_ingredients":len_ingredients})


# Applying the function to add the recipe informations

df_filtered[["len_directions", "len_ingredients"]] = df_filtered.title.apply(lambda x: get_recipe_info(x))
```

## **Quick 22-Minute Recipe Recommendations**
For those days when you need a quick, healthy meal, here’s a recommendation engine that suggests recipes that take 22 minutes or less to prepare, while also allowing calorie and rating filters.


```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display


# Filter the DataFrame for '22-minute meals'
df_22 = df_final[df_final['22-minute meals'] == 1]

# Sort the DataFrame by the 'protein' column
df_22_sorted = df_22.sort_values(by='protein', ascending=False)

# Extract the 'title', 'protein', and 'calories' columns
data_22_weight_gain = df_22_sorted[['title', 'protein', 'calories']]


# Function to display recipe details
def show_recipe(recipe_title, output_area):
    recipe_info = get_recipe_info(recipe_title)
    output_area.clear_output()  # Clear previous output for other recipes
    with output_area:
        # Use HTML to display the full recipe without truncation
        full_output = f"""
        <h3>Recipe for {recipe_title}</h3>
        <h4>Directions:</h4>
        <p>{recipe_info['directions']}</p>
        <h4>Ingredients:</h4>
        <p>{recipe_info['ingredients']}</p>
        """
        display(widgets.HTML(value=full_output))

# Function to get recipe information based on title
def get_recipe_info(recipe_title):
    """
    Takes a recipe title and returns a list of 
    recipe directions and ingredients for the recipe.

    Args:
        recipe_title (str): The title of the recipe

    Returns:
        pd.Series: A Series containing the directions and ingredients
    """
    recipe_dic = [i for i in full_recipe if "title" in i.keys() and i["title"] == recipe_title]
    if len(recipe_dic) == 0:
        return pd.Series({"directions": None, "ingredients": None})
    else:
        recipe_dic = recipe_dic[0]
        directions = recipe_dic["directions"]
        ingredients = recipe_dic["ingredients"]
        return pd.Series({"directions": directions, "ingredients": ingredients})
    


# Function to filter the data based on calorie choice
def filter_by_calories(data_22_weight_gain, calorie_limit):
    return data_22_weight_gain[data_22_weight_gain['calories'] < calorie_limit]

# Display clickable buttons for each recipe title with protein and calorie count
def display_recipes(data_22_weight_gain):
    output_area = widgets.Output()

    for index, row in data_22_weight_gain.iterrows():
        # Create a button with a title (truncated to 20 characters) and protein & calorie info
        truncated_title = row['title'][:20] + '...' if len(row['title']) > 20 else row['title']
        display_text = f"{truncated_title} (Protein: {row['protein']}g, Calories: {row['calories']} kcal)"
        
        # Create a button
        button = widgets.Button(description=display_text, layout=widgets.Layout(width='auto'))
        
        # Set up the button to display recipe information when clicked
        button.on_click(lambda b, title=row['title']: show_recipe(title, output_area))
        
        # Display the button in the notebook
        display(button)
    
    # Display the output area where the recipe details will be shown
    display(output_area)

# Main function to display options and recipes based on calorie limit
def display_with_calorie_options(data_22_weight_gain):
    # Dropdown for selecting calorie range
    calorie_dropdown = widgets.Dropdown(
        options=[('All Recipes', None), ('Less than 500 Calories', 500), ('Less than 700 Calories', 700)],
        value=None,
        description='Calorie Filter:',
        style={'description_width': 'initial'}
    )

    # Output area for displaying filtered recipes
    filtered_output = widgets.Output()

    def on_calorie_change(change):
        filtered_output.clear_output()
        with filtered_output:
            calorie_limit = change['new']
            if calorie_limit is None:
                # Show all recipes if no limit is selected
                display_recipes(data_22_weight_gain)
            else:
                # Filter recipes based on calorie limit
                filtered_data = filter_by_calories(data_22_weight_gain, calorie_limit)
                display_recipes(filtered_data)

    # Attach the event handler to the dropdown
    calorie_dropdown.observe(on_calorie_change, names='value')

    # Display the dropdown and the output for filtered recipes
    display(calorie_dropdown)
    display(filtered_output)

# Run the main function to display calorie filter and recipe options
display_with_calorie_options(data_22_weight_gain)
```

## **Longer Recipe Recommendations**
For those who enjoy cooking and have more time, here's a recommendation engine for longer, more elaborate meals. As with the quick recipes, this engine also allows filters for calories and star ratings.


```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output


# Function to filter the data based on rating, calorie, and limit
def filter_by_rating_and_calories(data, rating_threshold, calorie_limit, limit):
    filtered_data = data[data['rating'] >= rating_threshold]
    if calorie_limit is not None:
        filtered_data = filtered_data[filtered_data['calories'] < calorie_limit]
    filtered_data = filtered_data.sort_values(by='protein', ascending=False).head(limit)
    return filtered_data

# Function to get recipe information based on title
def get_recipe_info(recipe_title):
    recipe_dic = [i for i in full_recipe if "title" in i.keys() and i["title"] == recipe_title]
    if len(recipe_dic) == 0:
        return pd.Series({"directions": None, "ingredients": None})
    else:
        recipe_dic = recipe_dic[0]
        directions = recipe_dic["directions"]
        ingredients = recipe_dic["ingredients"]
        return pd.Series({"directions": directions, "ingredients": ingredients})

# Function to display recipe details
def show_recipe(recipe_title, output_area):
    recipe_info = get_recipe_info(recipe_title)
    output_area.clear_output()  # Clear previous output for other recipes
    with output_area:
        full_output = f"""
        <h3>Recipe for {recipe_title}</h3>
        <h4>Directions:</h4>
        <p>{recipe_info['directions']}</p>
        <h4>Ingredients:</h4>
        <p>{recipe_info['ingredients']}</p>
        """
        display(widgets.HTML(value=full_output))

# Display clickable buttons for each recipe title with protein, calorie, and rating count
def display_recipes(filtered_data):
    output_area = widgets.Output()

    for index, row in filtered_data.iterrows():
        truncated_title = row['title'][:20] + '...' if len(row['title']) > 20 else row['title']
        display_text = f"{truncated_title} (Protein: {row['protein']}g, Calories: {row['calories']} kcal, Rating: {row['rating']})"
        
        # Create a button
        button = widgets.Button(description=display_text, layout=widgets.Layout(width='auto'))
        
        # Set up the button to display recipe information when clicked
        button.on_click(lambda b, title=row['title']: show_recipe(title, output_area))
        
        # Display the button in the notebook
        display(button)
    
    # Display the output area where the recipe details will be shown
    display(output_area)

# Main function to display options for rating, calorie, and recipe limit
def display_with_rating_and_limit_options(data_22_weight_gain):
    # Dropdown for selecting rating
    rating_dropdown = widgets.Dropdown(
        options=[(f'{i} Star and above', i) for i in range(1, 6)],
        value=4,  # Default value is 4-star and above
        description='Rating Filter:',
        style={'description_width': 'initial'}
    )

    # Slider for selecting the number of recipes to show
    limit_slider = widgets.IntSlider(
        value=10,  # Default value is 10
        min=1, 
        max=20,  # Max limit for recipes to show
        step=1,
        description='Recipe Limit:',
        style={'description_width': 'initial'}
    )

    # Dropdown for selecting calorie range
    calorie_dropdown = widgets.Dropdown(
        options=[('All Recipes', None), ('Less than 500 Calories', 500), ('Less than 700 Calories', 700)],
        value=None,  # Default to no filter (show all)
        description='Calorie Filter:',
        style={'description_width': 'initial'}
    )

    # Output area for displaying filtered recipes
    filtered_output = widgets.Output()

    def update_recipes(change):
        filtered_output.clear_output()
        with filtered_output:
            rating_threshold = rating_dropdown.value
            limit = limit_slider.value
            calorie_limit = calorie_dropdown.value
            # Filter recipes based on rating threshold, calorie limit, and recipe limit
            filtered_data = filter_by_rating_and_calories(data_22_weight_gain, rating_threshold, calorie_limit, limit)
            display_recipes(filtered_data)

    # Attach the event handler to the dropdowns and slider
    rating_dropdown.observe(update_recipes, names='value')
    limit_slider.observe(update_recipes, names='value')
    calorie_dropdown.observe(update_recipes, names='value')

    # Display the rating dropdown, recipe limit slider, calorie dropdown, and the output for filtered recipes
    display(rating_dropdown)
    display(limit_slider)
    display(calorie_dropdown)
    display(filtered_output)

# Run the main function to display rating, calorie, and limit filter options
display_with_rating_and_limit_options(df_final)
```
